```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
pd.options.display.max_columns = None
pd.set_option('display.float_format', lambda x: '%.7f' % x)
```

```python
df1 = pd.read_csv("../../reports/results/old_experiments_10_epochs.csv")
df1["Train RMSE"] = np.sqrt(1000000 * df1["Train MSE"])
df1["Case 1 RMSE"] = np.sqrt(1000000 * df1["Case 1 MSE"])
df1["Case 2 RMSE"] = np.sqrt(1000000 * df1["Case 2 MSE"])
df1["Case 3 RMSE"] = np.sqrt(1000000 * df1["Case 3 MSE"])
```

```python
df1.groupby(['Name'])['Case 1 RMSE', 'Case 2 RMSE','Case 3 RMSE'].agg(['mean', 'std', 'max', 'min'])
```
